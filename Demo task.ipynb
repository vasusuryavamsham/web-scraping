```python
# Importing the packages

import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.yelp.com/search?cflt=restaurants&find_loc=San+Francisco%2C+CA"
```

```python
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.content,"html.parser")
print(soup.prettify())
```

```python
names = soup.find_all('a',class_="css-1m051bw",role = "link")
# names
```

```python
res_names = []
for i in names:
    res_names.append(i.text)
    
res_names.pop(0)
res_names
```

```python
address = []
add_tag = soup.find_all('p',class_="css-dzq7l1")
for i in add_tag:
    address.append(i.find_all('span',{'class':'css-chan6m'}))
address
```

```python
res_address = []
for i in address:
    res_address.append(i[0].text)
res_address
```

```python
res_ratings = []
ratings = soup.find_all('div',class_="five-stars__09f24__mBKym five-stars--regular__09f24__DgBNj display--inline-block__09f24__fEDiJ border-color--default__09f24__NPAKY")

for i in ratings:
    aria_label = i['aria-label']
    res_ratings.append(aria_label)

res_ratings    
```

```python
res_reviews = []
reviews = soup.find_all('span',class_="css-chan6m")
for i in range(0,20,2):
    res_reviews.append(reviews[i].text)

res_reviews
```

```python
d = { "Restaurent name":res_names,"Rating":res_ratings,"Reviews":res_reviews,"Address":res_address}
```

```python
#Creating DataFrame 

scraped_data = pd.DataFrame(d)
print(scraped_data)
```
